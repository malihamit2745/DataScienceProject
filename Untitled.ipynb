```python
to_remove = ['Weight', 'Duration']
df.drop(to_remove, axis=1, inplace=True)
```

```python
features = df.drop(['User_ID', 'Calories'], axis=1)
target = df['Calories'].values

X_train, X_val,\
    Y_train, Y_val = train_test_split(features, target,
                                      test_size=0.1,
                                      random_state=22)
X_train.shape, X_val.shape
features.head()
```

```python
# Normalizing the features for stable and fast training.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Define models
models = {
    "Linear Regression": LinearRegression(),
    "XGBoost Regressor": XGBRegressor(),
    "Lasso Regression": Lasso(),
    "Random Forest Regressor": RandomForestRegressor(),
    "Ridge Regression": Ridge()
}

# Dictionary to store results
results = {}

# Train & Evaluate each model
for name, model in models.items():
    model.fit(X_train, Y_train)  # Train model
    preds = model.predict(X_val)  # Make predictions
    
    # Compute evaluation metrics
    r2 = r2_score(Y_val, preds)
    mae = mean_absolute_error(Y_val, preds)
    mse = mean_squared_error(Y_val, preds)
    rmse = np.sqrt(mse)
    
    # Store results
    results[name] = {"R Score": r2, "MAE": mae, "MSE": mse, "RMSE": rmse}

# Print results
for model, metrics in results.items():
    print(f"\nModel: {model}")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

# Define the model
xgb = XGBRegressor(objective='reg:squarederror', random_state=42)

# Define hyperparameters to tune
param_grid = {
    'n_estimators': [100, 300, 500],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 6, 9]
}

# Use GridSearchCV for tuning
grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=5, scoring='r2', verbose=2, n_jobs=-1)
grid_search.fit(X_train, Y_train)

# Best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
# Train with best parameters
best_xgb = XGBRegressor(**best_params, objective='reg:squarederror', random_state=42)
best_xgb.fit(X_train, Y_train)

# Make predictions
preds = best_xgb.predict(X_val)

# Evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

r2 = r2_score(Y_val, preds)
mae = mean_absolute_error(Y_val, preds)
mse = mean_squared_error(Y_val, preds)
rmse = np.sqrt(mse)

# Print performance
print("\nFinal XGBoost Model Performance:")
print(f"R Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
import joblib
import pandas as pd

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees
    'learning_rate': [0.01, 0.05, 0.1],  # Step size shrinkage
    'max_depth': [3, 5, 7],  # Maximum depth of trees
    'subsample': [0.8, 1],  # Fraction of samples used for training each tree
    'colsample_bytree': [0.8, 1]  # Fraction of features used per tree
}

# Assume xgb is your XGBRegressor
xgb = XGBRegressor()

# Use GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(xgb, param_grid, cv=5, scoring='r2', n_jobs=-1, verbose=2)
grid_search.fit(X_train, Y_train)

# Best parameters from tuning
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Train the best model
best_xgb = XGBRegressor(
    n_estimators=best_params['n_estimators'],
    learning_rate=best_params['learning_rate'],
    max_depth=best_params['max_depth'],
    subsample=best_params['subsample'],
    colsample_bytree=best_params['colsample_bytree'],
    objective='reg:squarederror',  
    random_state=42
)

# Train the model
best_xgb.fit(X_train, Y_train) 

# Save the trained model
joblib.dump(best_xgb, 'calories_burnt_model.pkl')

# Load the trained model
loaded_model = joblib.load('calories_burnt_model.pkl')

# Define a sample input for prediction (Make sure the features match the model's training data)
sample_data = pd.DataFrame({
    'Gender': [2],
    'Age': [25],
    'Height': [169],  # in cm
    'Heart_Rate': [120],  # BPM
    'Body_Temp': [37.0]  # in Celsius
})

# Ensure the sample data columns are in the correct order and match the model's input
# Prediction
prediction = loaded_model.predict(sample_data)

print(f"Predicted Calories Burnt: {prediction[0]:.2f} kcal")
```
